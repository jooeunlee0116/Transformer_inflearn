# 판다스 기초문법

- 열을 접근하는 법 [], [[]]
- 행을 접근하는 법
    - loc - Location
        - 인덱싱 로케이션 처음 생성된 데이터프레임의 인덱스 기준으로 접근하는 것
        - 컬럼을 명시적으로 지정해서 사용할 수 있다. e.g. tt['survived']
    - iloc - Integer Location
        - 넘버링된 인덱스 -> 현재 기준의 데이터프레임의 순서 인덱스로 접근하는 것
        - 컬럼도 모두 다 int로 작성해야 한다.
- 2차원이므로 행과 열을 같이 추출하려면?
    - 데이터프레임.loc[:,:] - [행,열]
    - 데이터프레임.iloc[:,:]- [행,열]
        - loc, iloc 둘의 인덱싱 차이
            - loc -> 10:20 을 출력하면 10:20 모두 출력
            - iloc -> 10:20 을 출력하면 10:19 출력이 된다.
- shape - 행,열
- info() - 데이터프레임 정보
- head(), tail() - 앞에 데이터 확인, 뒤에 데이터 확인
- describe()- 기초통계수치 확인
- dtypes- 데이터 타입 확인
- columns - 컬럼들 확인
- index -인덱스 확인

## 상황에 따른 데이터프레임 조작
- 데이터 프레임 만드는 법
    - pd.DataFrame({열1:행값들1, 열2:행값들2...n})

- 변수명 바꾸는 법
    - df.rename(columns = {'기존컬럼':'변경할 컬럼'})
    
- 데이터프레임의 원본을 접근하기 위해서는 두 가지 방법을 사용
    - **copy() 데이터 원본 복제**
    - inplace=True 원본에 적용시키는 것
    
- 파생변수들 만들기!
    - 파생변수란?
        - 기존의 데이터를 활용해서 새로운 변수를 만드는 것
            - 시험점수 영어, 수학, 과학 = > (영어 +수학+ 과학) / 3 평균 계산
    - 파생변수를 직관적으로 데이터프레임에서 만들 수 있다.
    - assign, lambda 함수등을 배우겠지만 직관적 사용도 가능하다!
- 내가 원하는 데이터를 찾고, 내가 원하는 파생변수를 만들기!
    - numpy와 함께 사용하면 더욱 쉽게 찾을 수 있다.
    - numpy에서 제공하는 np.where 함수를 사용해서 특정 조건의 데이터를 찾을 수 있다.
        - 해당 조건을 활용해서 파생변수도 만들 수 있다.

```python
## numpy로 찾기
## 타이타닉 데이터 기준으로 나이가 20>= 이런 사람들을 찾고 싶다.
np.where(tt['age']>=20)


#where함수 사용하여 조건에 만족하는 인덱스 찾기
#tt['age']>=20 : age 열 값들이 20이상인지 확인하는 불리언 배열
#np.where: 불리언 배열에서 True인 인덱스 반환
```

```python
# 20살 이상은 '생존확률이 낮음' , 20살이 미만 '생존확률 높음'
tt['new_limit_age']=np.where(tt['age']>=20,'생존확률 낮음','생존확률 높음')

#데이터프레임에 새로운 열 추가
#조건이 True(age가 20 이상)인 경우 '생존확률 낮음' 문자열 할당
#조건이 False(age가 20 미만)인 경우 '생존확률 높음' 문자열 할당
```

## 데이터 분석에 필요한 기초 문법들
- query()
    - 성별은 여성인데, pclass 3에 있는 사람들의 fare 평균과 총합을 알고 싶어!
    - 여러 요구조건의 필요한 데이터를 추출하기 위한 기초 문법!
    - query("요구조건 == 값 "), 2개 이상의 조건은 query("요구조건1==값1 & 요구조건2==값2")
- 불린 인덱싱 df[df[True, Fales]]
- sort_values()
- groupby(), agg
- assign()
- merge(), concat() - 1시간 정도 따로 다룰 예정
- melt(), pivot()
- apply()


```python
## pclass 가 3인 데이터 and 조건으로 여성인 사람을 조건 추출
tt.query("pclass==3 & sex =='female'")

#query: 여러 요구 조건을 문자열로 표현하여 추출
```

```python
tt.query("pclass==3 & sex =='female'")['fare'].sum()

#조건에 해당하는 데이터프레임에서 fare 열의 합계 계산
```

```python
tt.query("pclass==3 & sex =='female'")['fare'].mean()


#조건에 해당하는 데이터프레임에서 fare 열의 평균 계산
```
